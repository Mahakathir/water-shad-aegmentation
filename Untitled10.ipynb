```python
import cv2 as a
import numpy as n
from matplotlib import pyplot as p
```

```python
img=a.imread("bel.jpg")
p.imshow(img)
```

```python
img_rgb=a.cvtColor(img,a.COLOR_BGR2RGB)
img_gray=a.cvtColor(img,a.COLOR_RGB2GRAY)
p.imshow(img_gray)
```

```python
img1m=a.medianBlur(img_gray,3)
ath2=a.adaptiveThreshold(img1m,255,a.ADAPTIVE_THRESH_GAUSSIAN_C,\
                         a.THRESH_BINARY,141,5)
p.imshow(ath2)
```

```python
kernal=n.ones((3,3),n.uint8)
opening=a.morphologyEx(ath2,a.MORPH_OPEN,kernal,iterations=5)
p.imshow(opening)
```

```python
sure_bg=a.dilate(opening,kernal,iterations=4)
p.imshow(sure_bg)
```
